```python
import matplotlib.pyplot as plt
import numpy as np
```

# type of graphs
1. plot - line graph , between two parameters
2. bar - bar graph of a single parameter and 2 both (optional: catagorical data )
3. hist - histogram of a bunch of data that is i.e. frequency of single parameter 
4. scatter - dot graph between two parameters
5. boxplot/box - staistical overview in form of box for a single parameter, it can be mean , outliner, qunatiles(25%,75%), IQR (optional: catagorical data)
6. stem - more like bar graph , lines insted of bars
7. violinplot - voilin shape statistical overview graph.
                single parameter (and optional: catagorical data). we can also get spread of data(more like histograph)

```python
x = np.linspace(0,5,10) # it decides the gap we give number of values
x1=np.arange(1,11,3) # it decides number of values we give the gap
```

```python
x
```

```python
x1
```

```python
y=x**3
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.bar(x,y)
```

```python
plt.hist(x)
plt.title('histograph')
```

```python
plt.scatter(x,y)
plt.grid()
plt.xlabel('x-->x')
plt.ylabel('y-->y')
plt.xticks()
plt.yticks(range(0,200,10))
plt.show()
```

```python
plt.boxplot(y)
```

```python
plt.stem(x,y)
```

```python
plt.violinplot(x)
plt.title('violin on x')
plt.show()
plt.violinplot(y)
plt.title('violin on y')
plt.show()
```

```python
y2=x**4
plt.plot(x,y,x,y2)
plt.legend(['y=x**3','y2=x**4'])
plt.show()
```

```python
fig = plt.figure(figsize=(10,5))
plt.plot(x,y, lw=3, marker='o', markersize=10, markerfacecolor='red')
plt.plot(x,y2, lw=3, marker='v',  markersize=10, markerfacecolor='yellow')
plt.legend(['y=x**3','y2=x**4'])
plt.grid()
plt.show()
```
