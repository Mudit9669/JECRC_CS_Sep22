1. plot - simple line plot
2. plot.area - Filled version of line plot
3. plot.bar / plot.barh - Bar graph
4. plot.density - Density Line
5. plot.box - Box plot
6. plot.hexbin - haxagon version of scatter with density
7. plot.hist - histogram
8. plot.kde - kernal density Estimator [Density Graph]
9. plot.pie - Pie chart of Catagorical Data

# Categories of functions
# 1. Catagorical plots (hue can be used | catagorical data)
    a. countplot (single parameter is used)
    b. barplot (one parameter should be catagorical and one should be numerical and it calculates bean by default)
    c. boxplot
    d. violinplot
    e. swarmplot
# 2. Distribution plots
    a. distplot
    b. displot
    c. jointplot
    d. pairplot
    e. kdeplot
    f. rugplot
    g. scatterplot
# 3. Matrix plots
    a. heatmap
    b. clustermap
    c. paiplot (same as above one)
# 4. Regression plots
    a. lmplot
    b. regplot
# 5. grid plots
    a. PairGrid
    b. JointGrid
    c. FacetGrid
