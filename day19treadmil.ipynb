#### Univariate Analysis

```python
# Catagorical Data
for col in cat_cols:
    plt.figure(figsize=(10,3))
    plt.subplot(1,2,1)
    sns.countplot(data=data, x=col , palette=choice(p_list))
    plt.title(col)
    plt.subplot(1,2,2)
    data[col].value_counts().plot.pie(autopct='%d%%', figsize=(8,8))
    plt.show()
```

```python
# Numerical Data
num_cols=['Age','Fitness','Usage','Education','Income','Miles']
```

```python
# Numerical Data
for col in num_cols:
    fig = plt.figure(figsize=(10,3))
    fig.suptitle(col)
    color=[random(),random(),random()]
    plt.subplot(1,3,1)
    sns.distplot(data[col], color = color)
    plt.grid()
    plt.subplot(1,3,2)
    sns.boxplot(data=data, x=col, color = color)
    plt.grid()
    plt.subplot(1,3,3)
    sns.violinplot(data=data, x=col, color = color)
    plt.grid()
    plt.show()
```

```python
for col in cat_cols:
    for ncol in num_cols:
        plt.figure(figsize=(10,3))
        plt.suptitle(f'{col}vs{ncol}')
        plt.subplot(1,3,1)
        sns.boxplot(data=data,x=col,y=ncol,palette=choice(p_list))
        plt.subplot(1,3,2)
        sns.violinplot(data=data,x=col,y=ncol,palette=choice(p_list))
        plt.subplot(1,3,3)
        sns.barplot(data=data,x=col,y=ncol,palette=choice(p_list))
        plt.show()
```

```python
pg=sns.PairGrid(data=data[num_cols],palette=choice(p_list))
pg.map_diag(plt.hist)
pg.map_lower(sns.regplot)
pg.map_upper(sns.kdeplot)
plt.grid()
```

```python
sns.heatmap(data=data[num_cols].corr(),annot=True,cmap=choice(p_list))
```

# Multivariate Analysis

```python
for col in cat_cols:
    if col=='Product':
        continue
    for ncol in num_cols:
        plt.figure(figsize=(10,3))
        plt.suptitle(f'{col}vs{ncol}')
        plt.subplot(1,3,1)
        sns.boxplot(data=data,x=col,y=ncol,hue='Product',palette=choice(p_list))
        plt.subplot(1,3,2)
        sns.violinplot(data=data,x=col,y=ncol,palette=choice(p_list),hue='Product')
        plt.subplot(1,3,3)
        sns.barplot(data=data,x=col,y=ncol,palette=choice(p_list),hue='Product')
        plt.show()
```
